# Hypothesis1: Technology products have the highest profit margin compared to other product categories.

```python
data.columns
```

```python
cat_profit=data.groupby('Category')['Profit'].sum()
cat_profit
```

```python
cat_profit.plot(kind='bar')
plt.title('Profit by Category')
plt.xlabel('Category')
plt.ylabel('Total Profit')
plt.show()
```

```python
#Conclusion 1: The hypothesis is supported as technology products have the highest profit margin of the three categories
```

# Hypothesis 2:The East region has the highest sales compared to other regions

```python
data.columns
```

```python
Reg_sales=data.groupby('Region')['Sales'].sum()
Reg_sales
```

```python
Reg_sales.plot(kind='bar', color='green')
plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```

# Hypothesis 3:Sales are higher during certain months of the year.

```python
data.head()
```

```python
data.columns
```

```python
data['Order Date']=pd.to_datetime(data['Order Date'])
data['Month']=data['Order Date'].dt.month
data
```

```python
Month_sales=data.groupby('Month')['Sales'].sum()
Month_sales.plot(kind='line', color='purple')
plt.title('Monthly Sales')
plt.xlabel('Months')
plt.ylabel('Total Sales')
plt.show()
```

```python
#Conclusion 3:The hypothesis is supported as sales are higher during certain months of the year.
```

# Hypothesis 4: Orders with the same-day shipping have the lowest rate of returned products.

```python
data.columns
```

```python
data['Ship Mode'].unique()
```

```python
total_orders_by_shipping_mode=data.groupby('Ship Mode').size()

returned_orders_by_shipping_mode=data[data['Profit']<0].groupby('Ship Mode').size()

returned_per_shipping_mode=(returned_orders_by_shipping_mode/total_orders_by_shipping_mode)*100
returned_per_shipping_mode
```

```python
returned_per_shipping_mode.plot(kind='bar', color='grey')
plt.title('Percentage Return by Shipping Mode')
plt.xlabel('Shipping Mode')
plt.ylabel('Percentage Return')
plt.show()
```

```python
#Conclusion 4:The hypothesis is supported as orders with the same-day shipping held the lowest rate of returned products
```

# Hypothesis 5: The company's profit is more on weekdays than on weekends.

```python
data.columns
```

```python
data['Order Date']=pd.to_datetime(data['Order Date'])
data['Day']=data['Order Date'].dt.day_name()

day_profit=data.groupby('Day')['Profit'].sum()
day_profit
```

```python
day_profit.plot(kind='bar', color='red')
plt.title('Total Profit by Day of week')
plt.xlabel('Day of the week')
plt.ylabel('Total profit')
plt.show()
```

```python
#Conclusion 5: The hypothesis is supported as company's profit is higher on weekdays as compared to weekends.
```
